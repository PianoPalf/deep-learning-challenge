# Deep Neural Network Model and Optimisations
## Accuracy and Loss Visualisation

```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set DPI of all Plots
plt.rcParams["figure.dpi"] = 150

# Set Dark Grid for all Plots
sns.set_style('darkgrid')
```

```python
# Import CSV Files
nn_model_df = pd.read_csv("Resources/AlphabetSoupCharity.csv")
optimisation_1_df = pd.read_csv("Resources/optimisation_1.csv")
optimisation_2_df = pd.read_csv("Resources/optimisation_2.csv")
optimisation_3_df = pd.read_csv("Resources/optimisation_3.csv")
```

```python
# Check optimisation_1_df
optimisation_1_df
```

```python
# Check optimisation_2_df
optimisation_2_df
```

```python
# Check optimisation_3_df
optimisation_3_df
```

```python
# Plot Accuracy and Loss for all Models using Subplots
fig, ax = plt.subplots(2, 4, figsize=(12, 5))

# Min and Max y-axis values for Accuracy
y_min_acc = 0.72
y_max_acc = 0.775

# Min and Max y-axis values for Loss
y_min_loss = 0.465
y_max_loss = 0.575

# Original Neural Network Model
# Define Accuracy subplot 
ax[0,0].plot(nn_model_df.index, nn_model_df['accuracy'], '#23b0bd')
ax[0,0].set_ylabel('accuracy')
ax[0,0].set_title("Neural Network Model", fontsize=10)
ax[0,0].set_xticklabels([])
ax[0,0].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot 
ax[1,0].plot(nn_model_df.index, nn_model_df['loss'], '#f65b74')
ax[1,0].set_xlabel('epoch')
ax[1,0].set_ylabel('loss')
ax[1,0].set_ylim(y_min_loss, y_max_loss)


# Optimisation 1
# Define Accuracy subplot
ax[0,1].plot(optimisation_1_df.index, optimisation_1_df['accuracy'], '#23b0bd')
ax[0,1].set_title("Optimisation 1", fontsize=10)
ax[0,1].set_yticklabels([])
ax[0,1].set_xticklabels([])
ax[0,1].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot
ax[1,1].plot(optimisation_1_df.index, optimisation_1_df['loss'], '#f65b74')
ax[1,1].set_xlabel('epoch')
ax[1,1].set_yticklabels([])
ax[1,1].set_ylim(y_min_loss, y_max_loss)


# Optimisation 2
# Define Accuracy subplot
ax[0,2].plot(optimisation_2_df.index,optimisation_2_df['accuracy'], '#23b0bd')
ax[0,2].set_yticklabels([])
ax[0,2].set_xticklabels([])
ax[0,2].set_title("Optimisation 2", fontsize=10)
ax[0,2].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot
ax[1,2].plot(optimisation_2_df.index, optimisation_2_df['loss'], '#f65b74')
ax[1,2].set_xlabel('epoch')
ax[1,2].set_yticklabels([])
ax[1,2].set_ylim(y_min_loss, y_max_loss)


# Optimisation 3
# Define Accuracy subplot
ax[0,3].plot(optimisation_3_df.index, optimisation_3_df['accuracy'], '#23b0bd', label='accuracy')
ax[0,3].set_yticklabels([])
ax[0,3].set_xticklabels([])

ax[0,3].set_title("Optimisation 3", fontsize=10)
ax[0,3].set_ylim(y_min_acc, y_max_acc)
ax[0,3].legend(loc='lower right')

# Define Loss subplot
ax[1,3].plot(optimisation_3_df.index, optimisation_3_df['loss'], '#f65b74', label='loss')
ax[1,3].set_xlabel('epoch')
ax[1,3].set_yticklabels([])
ax[1,3].set_ylim(y_min_loss, y_max_loss)
ax[1,3].legend()


# Set main title (super title)
plt.suptitle("Neural Network Models", fontsize=13, fontweight='bold')

# Save plot
plt.savefig('Resources/Images/nn_models_composite_plot.png')

# Show figure
plt.show()
```

```python
# Plot Accuracy and Loss for all Models using Subplots (One Plot)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Set Color palette
color = ['#f8cd82', '#f65b74', '#f72078', '#23b0bd', '#0df7db']

# Min and Max y-axis values for Accuracy
y_min_acc = 0.72
y_max_acc = 0.775

# Min and Max y-axis values for Loss
y_min_loss = 0.465
y_max_loss = 0.575

# Original Neural Network Model
# Define Accuracy subplot 
ax[0].plot(nn_model_df.index, nn_model_df['accuracy'], color[0], label='NN Model')
ax[0].set_ylabel('accuracy')
ax[0].set_xlabel('epochs')
ax[0].set_title('Accuracy')
ax[0].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot 
ax[1].plot(nn_model_df.index, nn_model_df['loss'], color[0], label='NN Model')
ax[1].set_xlabel('epochs')
ax[1].set_ylabel('loss')
ax[1].set_title('Loss')
ax[1].set_ylim(y_min_loss, y_max_loss)


# Optimisation 1
# Define Accuracy subplot
ax[0].plot(optimisation_1_df.index, optimisation_1_df['accuracy'], color[1], label='Optimisation 1')
ax[0].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot
ax[1].plot(optimisation_1_df.index, optimisation_1_df['loss'], color[1], label='Optimisation 1')
ax[1].set_ylim(y_min_loss, y_max_loss)


# Optimisation 2
# Define Accuracy subplot
ax[0].plot(optimisation_2_df.index,optimisation_2_df['accuracy'], color[3], label='Optimisation 2')
ax[0].set_ylim(y_min_acc, y_max_acc)

# Define Loss subplot
ax[1].plot(optimisation_2_df.index, optimisation_2_df['loss'], color[3], label='Optimisation 2')
ax[1].set_ylim(y_min_loss, y_max_loss)


# Optimisation 3
# Define Accuracy subplot
ax[0].plot(optimisation_3_df.index, optimisation_3_df['accuracy'], color[4], label='Optimisation 3')
ax[0].set_ylim(y_min_acc, y_max_acc)
ax[0].legend(loc='lower right')

# Define Loss subplot
ax[1].plot(optimisation_3_df.index, optimisation_3_df['loss'], color[4], label='Optimisation 3')
ax[1].set_ylim(y_min_loss, y_max_loss)
ax[1].legend()


# Set main title (super title)
plt.suptitle("Neural Network Models", fontsize=13, fontweight='bold')

# Save plot
plt.savefig('Resources/Images/nn_models_composite_plot_one.png')

# Show figure
plt.show()
```
